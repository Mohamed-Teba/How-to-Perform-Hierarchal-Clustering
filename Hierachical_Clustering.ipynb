```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/LearnDataSci/glossary/main/data/protein.csv")
df.sample(5)
```

```python
data = df.iloc[:,1:10]
data.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

C1 = linkage(data, method="single", metric="euclidean")
C2 = linkage(data, method="complete", metric="euclidean")
C3 = linkage(data, method="average", metric="euclidean")
C4 = linkage(data, method="ward", metric="euclidean")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 10))
plt.subplot(2, 2, 1) , plt.title("single"), dendrogram(C1)
plt.subplot(2, 2, 2) , plt.title("complete"), dendrogram(C2)
plt.subplot(2, 2, 3) , plt.title("average"), dendrogram(C3)
plt.subplot(2, 2, 4) , plt.title("ward"), dendrogram(C4)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

cluster1 = fcluster(C1, 2, criterion = "maxclust")
cluster2 = fcluster(C2, 2, criterion = "maxclust")
cluster3 = fcluster(C3, 2, criterion = "maxclust")
cluster4 = fcluster(C4, 2, criterion = "maxclust")

print('Cluster 1 :', cluster1, "\n")
print('Cluster 2 :', cluster2, "\n")
print('Cluster 3 :', cluster3, "\n")
print('Cluster 4 :', cluster4)
```
